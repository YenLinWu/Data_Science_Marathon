1. 將所有轉為周資料
2. 將周資料的值取平均

````python
import numpy as np
import pandas as pd
````

````python
index = pd.date_range( '1/1/2021', periods=20, freq='D' )
series = pd.Series( range(20), index=index )
series
````

````python
# 1. 將所有轉為周資料

series.to_period( freq='W' )
````

````python
# 2. 將周資料的值取平均

series.resample('W').mean() 
````
