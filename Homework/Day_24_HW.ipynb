```python
# 載入套件
import numpy as np
import matplotlib.pyplot as plt
import os
os.environ['PROJ_LIB'] = r'C:\Users\YenLin\anaconda3\pkgs\proj4-5.2.0-ha925a31_1\Library\share'
from mpl_toolkits.basemap import Basemap
```

```python
# 創建基本地圖
# 當 "projection" 參數為“ortho"時，所得圖位地球儀截面 

# 設定圖形大小
plt.figure( figsize=(5,5) )

# 設定投影型態
m = Basemap( projection='ortho', resolution=None, lat_0=0, lon_0=123 )

# 設定圖形顏色，選用 NASA 寶石藍輸出 
m.bluemarble( scale=0.3 ) 

# 秀圖
plt.show()
```

```python
# 創建基本地圖
# 當 "projection" 參數為 "Mill" 時，所得平展位面

# 設定圖形大小
plt.figure( figsize=(16,12) )

# 設定投影型態
m = Basemap( projection='mill', resolution=None, lat_0=0, lon_0=123 )

# 設定圖形顏色，選用 NASA 寶石藍輸出 
m.bluemarble( scale=0.3 ) 

# 秀圖
plt.show()
```

```python
plt.figure( figsize=(16,12) )
m = Basemap( projection='cyl', lat_0=50, lon_0=0, lat_1=45, lon_1=255 )
m.drawcoastlines( linewidth=1 )
m.drawcountries( linewidth=1 )
m.fillcontinents( color='lightgray', alpha=0.5 )
m.shadedrelief()
plt.show()
```

```python
# 設置 Lambert 保形底圖。
# width=12000000、height=9000000
# 經緯度 lat_1=45.、lat_2=55、lat_0=50、lon_0=-107
# set resolution = None 跳過邊界數據集的處理
# projection='lcc'

# 設定圖形大小
plt.figure( figsize=(16,12) )

m = Basemap( width=12000000, height=9000000, 
             lat_1=45., lat_2=55, lat_0=50, lon_0=-107.,
             resolution=None, projection='lcc' )

# 選用 NASA 寶石藍輸出 
m.bluemarble( scale=0.5 )
 
# 秀圖
plt.show( )
```

```python
# "projection" 參數為 "Mill"
# 設定經緯座標
# 左下角的緯度 - llcrnrlat = -90；
# 左下角的經度 - llcrnrlon = -180；
# 右上角的緯度 - urcrnrlat = 90；
# 右上角的經度 -urcrnrlon = 180；
# 成圖設定 resolution

plt.figure( figsize=(16,12) )
m = Basemap( projection='mill',
             llcrnrlat = -90, llcrnrlon = -180,
             urcrnrlat = 90, urcrnrlon = 180 )

# 創建邊線
m.drawcoastlines()
# 畫出國家，並使用線寬為2 的線條生成分界線。
m.drawcountries( linewidth=2 )
# 畫出州界
m.drawstates(color='b')

plt.title( 'Basemap Tutorial' )
plt.show()
```
