題目:<br>
1.生成一個等差數列，首數為0，尾數為20，公差為1的數列。<br>
2.呈上題，將以上數列取出偶數。<br>
3.呈1題，將數列取出3的倍數。<br>

```python
import numpy as np
print( 'NumPy 版本: ', np.__version__ )
```

```python
#1.生成一個等差數列，首數為0，尾數為20，公差為1的數列。
ArithmeticSequence = np.arange(0,21)
ArithmeticSequence
```

```python
#2.呈上題，將以上數列取出偶數。
even = [ i for i in ArithmeticSequence if i%2==0 ]
even
```

```python
#3.呈1題，將數列取出3的倍數。
odd = [ i for i in ArithmeticSequence if i>0 and i%3==0 ]
odd
```
