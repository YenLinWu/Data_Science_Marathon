作業重點<br>
線性代數公式應用<br>
矩陣相乘維度需要對好，例如:2X3矩陣乘上3X5矩陣得到2X5矩陣

題目:<br>
array1 = np.array([[10, 8], [3, 5]]), 即表示 <br>
$$
\text{array1} = 
 \left[
 \begin{matrix}
   10 & 8 \\
   3 & 5
  \end{matrix}
  \right] 
$$

1. 運用上列array計算反矩陣，乘上原矩陣，並觀察是否為單位矩陣?
2. 運用上列array計算特徵值、特徵向量?
3. 運用上列array計算SVD?

```python
import numpy as np
print( 'NumPy 版本: ', np.__version__ )
```

```python
#1. 運用上列array計算反矩陣，乘上原矩陣，並觀察是否為單位矩陣?
array1 = np.array([[10, 8], [3, 5]])

inv_array1 = np.linalg.inv( array1 )
print( '反矩陣 = \n', inv_array1 )
```

```python
print( inv_array1 @ array1 )
print( array1 @ inv_array1 )
```

```python
#2. 運用上列array計算特徵值、特徵向量?

v, w = np.linalg.eig(array1)

print( '特徵值=', v, '\n' )
print( '特徵向量= \n', w )
```

```python
#3. 運用上列array計算SVD?

np.linalg.svd(array1)
```
