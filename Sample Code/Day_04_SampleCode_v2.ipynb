```python
# 載入 NumPy 套件
import numpy as np

# 檢查正確載入與版本
print(np)
print(np.__version__)
```

```python
# 相同大小的陣列運算

import numpy as np

a = np.array( [20,30,40,50] )
b = np.arange( 4 )

print(a + b)
print(a * b)
```

```python
# 常數與陣列運算

import numpy as np

a = np.array( [20,30,40,50] )

print(a - 2)
print(a / 10)
```

```python
# 不同大小的陣列運算

import numpy as np

data = np.array([[1, 2], [3, 4], [5, 6]])
ones_row = np.array([[1, 1]])

print(data + ones_row)
```

```python
# 陣列運算與容器運算的差異

a = np.array( [20,30,40,50] )
print(a + 1)

a = [20, 30, 40, 50]
b = []
for i in a:
    b.append(i+1)
print(b)
```

```python
# 用於陣列的矩陣運算

a = np.array([[1.0, 2.0], [3.0, 4.0]])
print(a)
y = np.array([[5.], [7.]])
print(y)
```

```python
print(a.transpose())
print(np.linalg.inv(a))
print(np.trace(a))
print(np.linalg.solve(a, y))
```
