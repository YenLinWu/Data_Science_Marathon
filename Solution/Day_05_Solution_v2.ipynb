### 1. 產生一個 1-11 的一維陣列，並且把 3-6 由正數變成負數。



```python
# 記得先 Import 正確的套件

import numpy as np
```

```python
Z = np.arange(11)+1

Z[(3 <= Z) & (Z < 7)] *= -1
print(Z)
```

### 2. 試著從一個隨機陣列中，找出比 0.5 大的數有幾個？



```python
A = np.random.rand(3, 6)

print(A[A > 0.5].size)
```
